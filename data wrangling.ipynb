## Simple merge on a key

```python
import pandas as pd

# Load your dataset
df = pd.read_csv(r"C:\Users\ramya\Downloads\netflix_titles.csv\netflix_titles.csv")

# Inspect first few rows
print(df.head())
```

```python
print("Running clean test...")

df.columns = df.columns.astype(str)
df.columns = df.columns.str.strip()
df.columns = df.columns.str.lower()

print(df.columns)
```

```python
df.columns = df.columns.astype(str)
df.columns = df.columns.str.strip()
df.columns = df.columns.str.replace(';', '')   # 👈 IMPORTANT
df.columns = df.columns.str.lower()
```

```python
print(df.columns)
```

```python
import pandas as pd
merged=pd.merge(df,df,on="title")
```

### Simple merge on show_id


```python
merged = pd.merge(df, df, on="show_id")
print(merged.head())
```

###  Inner merge

```python
merged = pd.merge(df, df, on="show_id", how="inner")
print(merged.head())
```

###  Left merge

```python
merged = pd.merge(df, df, on="show_id", how="left")
print(merged.head())
```

### Right merge

```python
merged = pd.merge(df, df, on="show_id", how="right")
print(merged.head())
```

### Outer merge

```python
merged = pd.merge(df, df, on="show_id", how="outer")
print(merged.head())
```

### Merge with suffixes

```python
merged = pd.merge(df, df, on="show_id", suffixes=("_A","_B"))
print(merged.head())
```

### Merge on multiple keys

```python
merged = pd.merge(df, df, on=["show_id","type"])
print(merged.head())
```

### Merge with indicator

```python
merged = pd.merge(df, df, on="show_id", how="outer", indicator=True)
print(merged.head())
```

### Merge with validate

```python
merged = pd.merge(df, df, on="show_id", validate="one_to_one")
print(merged.head())
```

###  Merge selected columns

```python
df1 = df[["show_id","title","release_year"]]
df2 = df[["show_id","director","country"]]
merged = pd.merge(df1, df2, on="show_id")
print(merged.head())
```

### Merge with index

```python
merged = pd.merge(df, df, left_index=True, right_index=True)
print(merged.head())
```

### Merge with condition

```python
merged = pd.merge(df[df["release_year"]>2015], df, on="show_id")
print(merged.head())
```

### Merge with different join keys


```python
merged = pd.merge(df, df, left_on="show_id", right_on="title", how="outer")
print(merged.head())
```

### Merge and select columns

```python
merged = pd.merge(df, df, on="show_id")[["show_id","title_x","title_y"]]
print(merged.head())
```

### Merge multiple DataFrames

```python
df1 = df[["show_id","title"]]
df2 = df[["show_id","director"]]
df3 = df[["show_id","country"]]
merged = df1.merge(df2,on="show_id").merge(df3,on="show_id")
print(merged.head())
```

## 1.2 Join 

### Simple join on index

```python
joined = df.set_index("show_id").join(df.set_index("show_id"), lsuffix="_A", rsuffix="_B")
print(joined.head())
```

### Left join

```python
df1.join(df2, how='left', lsuffix='_df1', rsuffix='_df2')
```

### Right Join

```python
df1.join(df2, how='right', lsuffix='_df1', rsuffix='_df2')
```

### Outer Join

```python
df1.join(df2, how='outer', lsuffix='_df1', rsuffix='_df2')
```

###  Inner Join

```python
df1.join(df2, how='inner', lsuffix='_df1', rsuffix='_df2')
```

###  Join with Suffixes (explicit)


```python
df1.join(df2, lsuffix='_left', rsuffix='_right')
```

### Join Multiple DataFrames

```python
df_temp = df1.join(df2, lsuffix='_df1', rsuffix='_df2')
df_result = df_temp.join(pd.DataFrame({'country':['US','UK','IN']}, index=[1,2,3]))
print(df_result)
```

 ### Join with Reindex

```python
df1.join(df2.reindex([1,2,3]), lsuffix='_df1', rsuffix='_df2')
```

### Join with Selected Columns


```python
df1.join(df2[['director']], lsuffix='_df1', rsuffix='_df2')
```

### Join with Overlapping Index

```python
df1.join(df2, how='outer', lsuffix='_df1', rsuffix='_df2')
```

### Join with Reset Index

```python
df1.reset_index().join(df2.reset_index(), lsuffix='_l', rsuffix='_r')
```

### Join with Multi-Index

```python
# Example: create a custom index of the same length
df1 = df1.set_index(pd.RangeIndex(start=1, stop=len(df1)+1))
df2 = df2.set_index(pd.RangeIndex(start=1, stop=len(df2)+1))

# Now join safely with suffixes
df1.join(df2, lsuffix='_df1', rsuffix='_df2')
```

### Join with Filtered Data

```python
df1.join(df2[df2['director']=='Dir2'], lsuffix='_df1', rsuffix='_df2')
```

### Join with Query

```python
df1.join(df2.query("director=='Dir1'"), lsuffix='_df1', rsuffix='_df2')
```

### Join with Sorted Index

```python
df1.join(df2.query("director=='Dir1'"), lsuffix='_df1', rsuffix='_df2')
```
